## Part 1: Importing the Data

### Instructions:
- Import the `pandas` library for dataframes.  Then `Matplotlib` and `Seaborn` for data visualization.
- Import data from the rent.csv into a dataframe from the tidytuesday link.
- Use `.info()` and `.head()` to inspect the structure and preview the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2022/2022-07-05/rent.csv'
df = pd.read_csv(url)
```

## Part 2: Inspecting, Cleaning & Removing Outliers

### Instructions:
- Identify missing values.
- Check data types are correct(e.g., dates, numeric columns).
- Check for outliers key numeric variables like `price`, `sqft`, `beds`, or `baths`.

### Think about:
- Are there missing values?
- Do the data types look appropriate?

```python
# Check for missing values
df.isnull().sum()
```

```python
# Basic summary statistics
df[['price', 'beds', 'baths', 'sqft']].describe()
```

```python
# Check data types
df.dtypes
```

```python
# Remove price outliers (1stâ€“99th percentile)
q_low = df['price'].quantile(0.01)
q_high = df['price'].quantile(0.99)
df = df[(df['price'] >= q_low) & (df['price'] <= q_high)]
```

## Part 3: Basic Exploration

Use `groupby` and `value_counts` to summarize trends across neighborhoods and cities.

```python
# Top cities by count
df['city'].value_counts().head(10)
```

## Part 4: Univariate Analysis

Explore individual variables to understand their distributions and frequency.

### Tasks:
- Plot histograms for numeric variables (`price`, `sqft`)
- Plot countplots for categorical variables (`beds`, `nhood`)


```python
# Histogram: Price
plt.hist(df['price'], bins=30)
plt.title("Distribution of Rental Prices")
plt.xlabel("Price ($)")
plt.ylabel("Count")
plt.show()
```

```python
# Histogram: Square Footage
plt.hist(df['sqft'].dropna(), bins=30)
plt.title("Distribution of Square Footage")
plt.xlabel("Sqft")
plt.ylabel("Count")
plt.show()
```

```python
# Bar plot of top 10 neighborhoods by number of listings
top_nhoods = df['nhood'].value_counts().head(10)

sns.barplot(x=top_nhoods.values, y=top_nhoods.index)
plt.title("Top 10 Neighborhoods by Listing Count")
plt.xlabel("Number of Listings")
plt.ylabel("Neighborhood")
plt.show()
```

```python
# Visual 1 ðŸ”§
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='baths', bins=20, kde=False, color='skyblue', edgecolor='black')

plt.title("Distribution of Bathrooms in Listings")
plt.xlabel("Number of Bathrooms")
plt.ylabel("Number of Listings")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

ðŸ”§ Visual 1 comment: We are able to see that most listing have 1 to 2 bathrooms and this makes up a majority of our dataset. Listing with 3 or more bathrooms are much less common, this shows us that likely smaller units take up much of the market

```python
# Visual 2 ðŸ”§
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

avg_sqft_by_bath = df.groupby('baths')['sqft'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=avg_sqft_by_bath, x='baths', y='sqft', palette='Blues_d')

plt.title("Average Square Footage by Number of Bathrooms")
plt.xlabel("Number of Bathrooms")
plt.ylabel("Average Square Footage")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

ðŸ”§ Visual 2 comment: Interpreting this visual shows us that there is a strong relationship between average square footage and number of bathrooms. We can see that starting at 1 bath room the average square footage is ~800, then as we go up in bath rooms the square footage increases at about ~300 square feet per additional bathroom.




## Part 5: Bivariate Analysis

Explore relationships between two variables to understand how features like square footage or bedrooms influence price.


```python
# Correlation matrix
corr_matrix = df[['price', 'beds', 'baths', 'sqft']].corr()
corr_matrix
```

```python
# Heatmap of correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Scatterplot: Square Footage vs Price
sns.scatterplot(x='sqft', y='price', data=df)
plt.title("Square Footage vs Price")
plt.xlabel("Square Feet")
plt.ylabel("Price ($)")
plt.show()
```

```python
# Average price by number of bedrooms
avg_price_beds = df.groupby('beds')['price'].mean().sort_index()
sns.barplot(x=avg_price_beds.index, y=avg_price_beds.values)
plt.title("Average Rental Price by Number of Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Average Price ($)")
plt.show()
```

### ðŸ”§ Try It Yourself â€“ Part 5

1. Create a scatterplot of `baths` vs `price`.  
2. Group by `year` and plot the average price over time.
3. What trends or outliers do you see?

```python
#1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='baths', y='price', alpha=0.3)

plt.title("Price vs. Number of Bathrooms")
plt.xlabel("Number of Bathrooms")
plt.ylabel("Price ($)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

```python
#2
import matplotlib.pyplot as plt
import seaborn as sns

avg_price_by_year = df.groupby('year')['price'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.lineplot(data=avg_price_by_year, x='year', y='price', marker='o')

plt.title("Average Rental Price Over Time")
plt.xlabel("Year")
plt.ylabel("Average Price ($)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

ðŸ”§ Add comment here: The scatterplot gives us some important details regarding overlap in the data, the graph is showing us that there are some 1 and 2 bathroom units that are as expensive as 4,5, even 6, bathroom units. The line chart shows us that prices remained somewhat stable from 2000 to 2007, from which they exploded in 2010. These two graphs show us that external factors like location and economic factors may be more important than just bathroom count.
